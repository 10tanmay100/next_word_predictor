```python
#reading a txt file
with open("/content/drive/MyDrive/CAMPUS X ML CODE/youtube.txt","r") as f:
  text=f.read()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer=Tokenizer()
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
#read linewise sentences and converting them into number or tokens
input_sq=[]
for sentence in text.split("\n"):
  tokenized_sentence=tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenized_sentence)):
    input_sq.append(tokenized_sentence[:i+1])
print(input_sq)
```

```python
#now as we can see lengths are different so we need to apply padding here.
#getting maximum word length
max_len=max([len(i) for i in input_sq])
#applying padding on prior to the numbers
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_tokens=pad_sequences(input_sq,maxlen=max_len,padding="pre")
padded_tokens
```

```python
#getting x and y
#so x will be all the numbers excdot the last one
#so y will be the last number of the array
X=padded_tokens[:,:-1]
y=padded_tokens[:,-1]
```

```python
#converting output thing into one hot encoded format
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=len(tokenizer.word_index)+1)
```

```python
X.shape
```

```python
#setup the model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, LSTM


#builiing the model
model = Sequential()
model.add(Embedding(len(tokenizer.word_index)+1, 900, input_length=330))
model.add(LSTM(units=300,input_shape=(330,1),return_sequences=True))
model.add(LSTM(units=300))
model.add(Dense(len(tokenizer.word_index)+1, activation='softmax'))
```

```python
model.compile(optimizer="rmsprop",loss="categorical_crossentropy",metrics=["accuracy"])
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
model.save("yt.h5")
```

```python
import wikipedia

def download_wikipedia_article(article_title, file_name='youtube.txt'):
    try:
        # Get the summary of the article
        summary = wikipedia.page(article_title).content

        # Save the content to a text file
        with open(file_name, 'w', encoding='utf-8') as file:
            file.write(summary)
            print(f"The content of the article '{article_title}' has been saved to '{file_name}'.")

    except wikipedia.exceptions.DisambiguationError as e:
        print(f"DisambiguationError: The article title '{article_title}' is ambiguous. Options: {e.options}")
    except wikipedia.exceptions.PageError:
        print(f"PageError: The page '{article_title}' does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
download_wikipedia_article('youtube')
```
